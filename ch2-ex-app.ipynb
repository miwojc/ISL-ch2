Create new qualitative variable "Elite" by binning "Top10perc" variable in two groups: >50% and <=50%

```r
Elite = rep("No",nrow(college))
Elite[college$Top10perc > 50]="Yes"
Elite=as.factor(Elite)
college=data.frame(college,Elite)
```

```r
head(college)
```

```r
summary(college)
```

Boxplot of "Outstate" vs "Elite"

```r
plot(college$Elite,college$Outstate,main="Boxplot Outstate versus Elite",xlab="Elite indicator",ylab="Out-of-state tuition")
```

Plot some histograms

```r
par(mfrow=c(2,2))
hist(college$Apps)
hist(college$Accept)
hist(college$Enroll)
hist(college$Top10perc)
hist(college$Top10perc,breaks=5, main="breaks=5")
hist(college$Top10perc,breaks=10, main="breaks=10")
hist(college$Top10perc,breaks=15, main="breaks=15")
hist(college$Top10perc,breaks=100, main="breaks=100")
```

Graduation rate vs Uni boxplot

```r
barplot(college$Grad.Rate, names.arg=rownames(college), xlab="Graduation rate", ylab="University")
```

Seems there is one Uni with Grad rate above 100%, this looks like mistake... let's add mistake column for Grad.Rate (gr_mistake) with Yes and No possible inputs

```r
gr_mistake = rep("No", nrow(college))
gr_mistake[college$Grad.Rate > 100] = "Yes"
gr_mistake = as.factor(gr_mistake)
college = data.frame(college, gr_mistake)
```

```r
summary(college)
```

Boxplot of grad rate vs Elite

```r
plot(college$Elite,college$Grad.Rate,main="Boxplot Graduation rate versus Elite",xlab="Elite indicator",ylab="Graduation rate")
```

Looks like mean of graduation rate is higher for "elite" students, aslo spread is lower for "Elite". Again we see that point above 100%

Instructional expenditure per student vs Graduation rate

```r
plot(college$Expend, college$Grad.Rate,main="Instructional expenditure per student vs Graduation rate",xlab="Instructional expenditure per student",ylab="Graduation rate")
```

Relationship doesn't seem to be linear. The more money spend on instructional expenditure doesn't necessarily mean higher graduation rate.
